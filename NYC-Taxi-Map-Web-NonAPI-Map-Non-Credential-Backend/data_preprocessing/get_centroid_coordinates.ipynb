```python
# read taxi_zones.shp geojson
import geopandas as gpd
import pandas as pd
```

```python
import geopandas as gpd

# Load the shapefile
taxi_zones = gpd.read_file("taxi_zones/taxi_zones.shp")

# Convert to a projected CRS (e.g., EPSG:3857 - Web Mercator, good for NYC)
taxi_zones_projected = taxi_zones.to_crs(epsg=3857)

# Compute centroids in the projected CRS
taxi_zones_projected['centroid'] = taxi_zones_projected['geometry'].centroid

# Convert back to geographic CRS (WGS84 - EPSG:4326)
taxi_zones = taxi_zones_projected.to_crs(epsg=4326)

# Convert centroid column to EPSG:4326 as well
taxi_zones['centroid'] = taxi_zones_projected['centroid'].to_crs(epsg=4326)
taxi_zones
```

```python
# Display the GeoDataFrame
taxi_zones = taxi_zones[['LocationID', 'centroid']]

results = []
for idx, row in taxi_zones.iterrows():
    results.append([row['LocationID'], row['centroid'].x, row['centroid'].y])

results
```

```python
# write results to csv
import csv
with open('taxi_zones_centroids.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["LocationID", "Longitude", "Latitude"])
    writer.writerows(results)
```
